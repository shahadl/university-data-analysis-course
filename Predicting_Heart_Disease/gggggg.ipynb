```r
#read the data
df <- read.csv(file="Video_Games_Sales_as_at_22_Dec_2016.csv",header=TRUE,sep=',')
head(df,5)
```

```r
df1=na.omit(df)
df1
```

```r
str(df1)
```

```r
#fit the Linear regression model
model1 <- lm(as.numeric(df1$User_Score)~NA_Sales+EU_Sales+JP_Sales+Other_Sales+Global_Sales,data <- df1)
```

```r
#display the summary of the model
summary(model1)
```

```r
#plot the regression line
ggplot(df1, aes(x = NA_Sales,EU_Sales,JP_Sales,Other_Sales,Global_Sales, y = as.numeric(df1$User_Score))) + 
  geom_point() +
  stat_smooth(method = "lm", col = "red")
```
